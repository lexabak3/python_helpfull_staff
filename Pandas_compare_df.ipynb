```python
import pandas as pd
import random
import string
```

```python
def mask_column(df, column, prefix='Type_'):
    # Get unique non-null values from the column
    unique_vals = df[column].dropna().unique()
    
    # Create a mapping from original to masked values
    masked_map = {
        val: f"{prefix}{i}" for i, val in enumerate(unique_vals, start=1)
    }
    return df[column].map(masked_map)


def compare_column_distribution(df1, df2, column):
    print(f"{len(df1)/len(df2) * 100:.2f}")
    
    dist1 = df1[column].value_counts(normalize=True).sort_index().rename('df1_%')
    dist1_2 = df1[column].value_counts().sort_index().rename('df1_qty')
    dist2 = df2[column].value_counts(normalize=True).sort_index().rename('df2_%')
    dist2_2 = df2[column].value_counts().sort_index().rename('df2_qty')    
    
    comparison = pd.concat([dist1,dist1_2,dist2,dist2_2], axis=1).fillna(0)

    comparison['diff_%'] = (comparison['df1_%'] - comparison['df2_%']).round(4)
    comparison['diff_qty'] = (comparison['df1_qty'] / comparison['df2_qty']).round(4)
    
    comparison[['df1_%', 'df2_%', 'diff_%', 'diff_qty']] = (comparison[['df1_%', 'df2_%', 'diff_%', 'diff_qty']] * 100).round(2)

    return comparison.sort_values('df2_qty',ascending=False)
```

```python
df = pd.read_csv('Data/Orders.csv')
df_subclass = df[df.feature == 1].copy()
```

```python
compare_column_distribution(df_subclass, df, 'Type')
```
